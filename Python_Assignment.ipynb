```python
x = [[12,7,3],
     [11,4,5],
    [11,23,45]]
y = [[12,7,3],
     [11,4,5],
    [11,23,45]]

res = [[0,0,0],[0,0,0],[0,0,0]]

#Adding two matrices
for i in range(len(x)):
    for j in range(len(x[1])):
        res[i][j] = x[i][j] + y[i][j]

for r in res:
    print(r)    
  
# transposing the matrix
for i in range(len(x)):
    for j in range(len(x[i])):
        res[j][i]=x[i][j]

for i in res:
    print(i)        
```
